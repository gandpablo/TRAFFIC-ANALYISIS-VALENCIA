```python
import pandas as pd
import numpy as np

def optimizar(df):

    df = df[['IdPM',
             'Descripcion','FechaHora',
             'Intensidad','Ocupacion','Velocidad',
             'FiabilidadIntensidad',
             'FiabilidadOcupacion',
             'FiabilidadVelocidad']]
    
    dic = df[['IdPM', 'Descripcion']].drop_duplicates().set_index('IdPM')['Descripcion'].to_dict()

    df['FechaHora'] = pd.to_datetime(df['FechaHora'])
    
    df['Año'] = df['FechaHora'].dt.year.astype('int16')     
    df['Mes'] = df['FechaHora'].dt.month.astype('int8')     
    df['Dia'] = df['FechaHora'].dt.day.astype('int8')       
    df['Hora'] = df['FechaHora'].dt.hour.astype('int8')     
    df['Minuto'] = df['FechaHora'].dt.minute.astype('int8') 
    df['DiaSemana'] = df['FechaHora'].dt.weekday + 1
    df['DiaSemana'] = df['DiaSemana'].astype('int8') 
    
    df = df.drop('FechaHora',axis=1)
    df = df.drop('Descripcion',axis=1)

    df['Intensidad'] = df['Intensidad'].astype('int16') 
    df['Ocupacion'] = df['Ocupacion'].astype('int16')
    df['Velocidad'] = df['Velocidad'].astype('int16')
    df['FiabilidadIntensidad'] = df['FiabilidadIntensidad'].astype('int16')
    df['FiabilidadOcupacion'] = df['FiabilidadOcupacion'].astype('int16')
    df['FiabilidadVelocidad'] = df['FiabilidadVelocidad'].astype('int16')
    
    return df,dic
```
